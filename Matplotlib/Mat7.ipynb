```python
#Data Visualization challenge
#Plots the length of movie title names on X-axis
#Frequently counts on Y-axis ie Number of movie having x character in thier title

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv("movie_metadata.csv")
df.head(n=10)

titles=list(df.get("movie_title"))


freq_title={

}

for title in titles:
    length=len(title)

    if freq_title.get(length) is None:
        freq_title[length]=1
    else:
        freq_title[length]+=1

X=np.array(list(freq_title.keys()))
Y=np.array(list(freq_title.values()))
print(X,Y)
plt.scatter(X,Y)
plt.show()
```
